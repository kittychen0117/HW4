```python
import datetime
import time

OneWeek = "https://query1.finance.yahoo.com/v7/finance/download/{}?period1={}&period2={}&interval=1wk&events=history&crumb=pVo323aD3D1"
OneMonth = "https://query1.finance.yahoo.com/v7/finance/download/{}?period1={}&period2={}&interval=1mo&events=history&crumb=pVo323aD3D1"


```

```python
def get_Oldestimestamp():  # get now timestamp and 3 years age
    
    nowtime = datetime.datetime.now().date().timetuple()
    nowyear, nowmonth, nowday = datetime.datetime.now().strftime("%Y-%m-%d").split('-')
    threeyeartoday = datetime.datetime.strptime("{}-{}-{}".format(str(int(nowyear)-3), nowmonth, nowday), "%Y-%m-%d")
    threeyeartoday = threeyeartoday.timetuple()
    return  int(time.mktime(nowtime)),  int(time.mktime(threeyeartoday))


now, past = get_Oldestimestamp()
(now - past) / 86400 / 365
```

```python
# First Get ALL Month , Weeks for dataframe column
now, past = get_Oldestimestamp()
WeekCol = []
MonthCol = []
garbage = ['', 'Date']
r=requests.post(OneWeek.format(ETF_List[0], past, now))
WeekCol=[d.split(',')[0] for d in r.text.split('\n')[1:] if d not in garbage ]
r=requests.post(OneMonth.format(ETF_List[0], past, now))
MonthCol=[d.split(',')[0] for d in r.text.split('\n')[1:] if d not in garbage ]
WeekCol[:5], MonthCol[:5]
```

```python
import numpy as np
import random
import time
import math
from scipy.stats import skew

random.seed(132912)
def CalRateofReturn(preprice, curprice):
    
    return float((curprice - preprice) / preprice)

def CalExcessReturn(RateofReturn, BondYield):
    
    return RateofReturn - BondYield


def caculateError(Bound, left, right):
    

    leftVal = np.sum(np.array([Bound**item for item in left]))
    rightVal = 1 * len(left)

    return (leftVal-rightVal)
    

def GetULBound(left, right):
    
    
    
    LB=random.random()
    RB=random.random()
    
    Lerror=caculateError(LB, left, right)
    Rerror=caculateError(RB, left, right)
    
    while(Lerror * Rerror > 0):
        RB = max(LB, RB)+random.random()
        LB = min(LB, RB)
        Lerror=caculateError(LB, left, right)
        Rerror=caculateError(RB, left, right)


    return RB, LB


def CalAlpha_g(ExcessReturnList):
    
    leftOperand=np.array(ExcessReturnList)
    RightOperand=np.array([0])
    
    
    upperBound, lowerBound = GetULBound(leftOperand, RightOperand)
    
    tolerance=0.00000001
    err=0
    cur = (upperBound+lowerBound)/2
    if caculateError(lowerBound, leftOperand, RightOperand) < 0:
        lowerBound, upperBound = upperBound, lowerBound
    while(abs(caculateError(cur, leftOperand, RightOperand)) > tolerance):
#         print(lowerBound, cur,upperBound,caculateError(lowerBound, leftOperand, RightOperand),caculateError(cur, leftOperand, RightOperand), caculateError(upperBound, leftOperand, RightOperand))
        if caculateError(cur, leftOperand, RightOperand) > 0:
            lowerBound=cur
            cur=(upperBound+lowerBound)/2
        elif caculateError(cur, leftOperand, RightOperand) < 0:
            upperBound=cur
            cur=(upperBound+lowerBound)/2

#     print(cur, caculateError(cur, leftOperand, RightOperand))
#     print("==========")
    return cur

def CalSR(RoR, r, t):
    
    RoR=np.array(RoR)
    mu=np.mean(RoR)
    sigma=np.std(RoR)
    
    sr=((mu-r)/sigma)*math.sqrt(t)
    return sr

def CalASSR(RoR, BondYield, flag):
    
    r=np.mean(np.array(BondYield))
    if flag == "WK":
        r=r/52
    else:
        r=r/12
    b3=2
    if flag == "WK":
        t=float(1)/52
    else:
        t=float(1)/12
    
    SR=CalSR(RoR, r, t)
    Skew=skew(RoR)
    
    ss=(SR*Skew*b3/3)+1
    return SR * (math.sqrt(ss))
    

def check(A_g, RoR):
    
    ans=len(RoR)
    s=0
    for g in RoR:
        s+=A_g**g
    print(ans, s)
    
    
    
    
    
```
